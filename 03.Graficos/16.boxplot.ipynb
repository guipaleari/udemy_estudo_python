```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Gerar BoxPlot do Volume:
plt.boxplot(base.Volume, vert=False, showfliers=True, notch=False, patch_artist=True) # ← Notch significa entalhe
plt.title('Boxplot com dados de Volume da base "Árvores"')
plt.xlabel('← Dados de Volume →')
```

```python
plt.boxplot(base.Volume, vert=False, showfliers=False, notch=True, patch_artist=True) # ← Notch significa entalhe
# ↑ Mesmo gráfico de BoxPlot do Volume com layout diferente:
plt.title('Boxplot com dados de Volume da base "Árvores"')
plt.xlabel('← Dados de Volume →')
```

```python
# Exibir gráfico boxplot de todos os dados:
plt.boxplot(base)
plt.title('Árvores')
plt.xlabel('Dados')
```

```python
# Gerar 3 gráficos diferentes de boxplots com cada um exibindo informações diferentes:
plt.boxplot(base.Volume, vert=False)
plt.boxplot(base.Girth, vert=False)
plt.boxplot(base.Height, vert=False)
plt.title('Árvores')
plt.xlabel('Dados')
```
