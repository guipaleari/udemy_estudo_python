```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
base1 = pd.read_csv('trees.csv')
base1.head()
```

```python
#Exibir histograma com 10 divisões (bins) e com gráfico de densidade (kernel density = suavização)
# stat: count, frequency, probability, percent, density
sns.histplot(base1.Volume, color='blue', kde=True, stat='density', bins=10).set(title='Dados de Volume')
```

```python
# Exibir o gráfico de Densidade considerando apenas o valor de 'horsebean'
sns.kdeplot(base2.loc[base2['feed'] == 'horsebean'].weight).set(title='Horsebean')
```

```python
# Exibir o Histograma considerando apenas o valor de 'casein'
sns.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True).set(title='Casein')
```

```python
# Exibir o Histograma considerando apenas o valor de 'linseed'
sns.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde=True).set(title='Linseed')
```

```python
# Exibir o Histograma considerando apenas o valor de 'meatmeal'
sns.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde=True).set(title='Meat Meal')
```

```python
# Exibir o Histograma considerando apenas o valor de 'soybean'
sns.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde=True).set(title='Soy Bean')
```

```python
# Exibir o Histograma considerando apenas o valor de 'sunflower'
sns.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde=True).set(title='Sunflower')
```

```python
# Exibir todos os gráficos em um só (3 linhas, 2 colunas):
plt.figure()

plt.subplot(3, 2, 1)
sns.histplot(base2.loc[base2['feed'] == 'horsebean'].weight, kde=True).set(title='horsebean')

plt.subplot(3, 2, 2)
sns.histplot(base2.loc[base2['feed'] == 'casein'].weight, kde=True).set(title='casein')

plt.subplot(3, 2, 3)
sns.histplot(base2.loc[base2['feed'] == 'linseed'].weight, kde=True).set(title='linseed')

plt.subplot(3, 2, 4)
sns.histplot(base2.loc[base2['feed'] == 'meatmeal'].weight, kde=True).set(title='meatmeal')

plt.subplot(3, 2, 5)
sns.histplot(base2.loc[base2['feed'] == 'soybean'].weight, kde=True).set(title='soybean')

plt.subplot(3, 2, 6)
sns.histplot(base2.loc[base2['feed'] == 'sunflower'].weight, kde=True).set(title='sunflower')

# Ajustar o layout para não ocorrer sobreposição:
plt.tight_layout
```
