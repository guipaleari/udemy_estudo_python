```python
import pandas as pd
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Visualização de um único BoxBlot em uma única imagem:
sns.boxplot(data=base.Volume, orient='v').set_title('Boxplot dos dados de volume da base "Árvores"')
```

```python
# Visualização de vários BoxBlot em uma única imagem:
sns.boxplot(data=base)
```
