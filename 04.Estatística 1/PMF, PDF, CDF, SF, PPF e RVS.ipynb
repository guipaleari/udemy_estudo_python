# **PMF, PDF, CDF e SF**

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import binom, norm
```

## **PMF (Probability Mass Function)**
* **Defini√ß√£o:** A PMF (Fun√ß√£o de Massa de Probabilidade) √© usada para distribui√ß√µes de probabilidade discretas. Ela fornece a probabilidade de uma vari√°vel aleat√≥ria discreta ser exatamente igual a algum valor.
* **Exemplo:** Considere o lan√ßamento de um dado. A PMF d√° a probabilidade de cada valor poss√≠vel (1, 2, 3, 4, 5, 6), que √© $1/6$ para cada valor.
* **F√≥rmula:** $ùëÉ(ùëã=ùë•)$, onde $ùëã$ √© uma vari√°vel aleat√≥ria discreta.

```python
# PMF para distribui√ß√£o binomial 
n = 10
p = 0.5
k = np.arange(0, n+1)
pmf = binom.pmf(k, n, p)

plt.subplot(2, 1, 1)
plt.bar(k, pmf)
plt.title('PMF - Distribui√ß√£o Binomial')
plt.xlabel('N√∫mero de Sucessos')
plt.ylabel('Probabilidade')
```

## **PDF (Probability Density Function)**
* **Defini√ß√£o:** A PDF (Fun√ß√£o Densidade de Probabilidade) √© usada para distribui√ß√µes de probabilidade cont√≠nuas. Diferente da PMF, a PDF descreve a densidade de probabilidade, n√£o a probabilidade direta. A probabilidade de a vari√°vel assumir um valor exato √© sempre zero; ao inv√©s disso, a PDF √© usada para encontrar a probabilidade de a vari√°vel cair em um intervalo espec√≠fico.
* **Exemplo:** Para vari√°veis cont√≠nuas como altura, a PDF pode ser usada para encontrar a probabilidade de a altura estar entre 150 e 160 cm.
* **F√≥rmula:** $ùëì(ùë•)$, onde $ùëã$ √© uma vari√°vel aleat√≥ria cont√≠nua e $ùëì(ùë•)$ √© a densidade de probabilidade.

```python
# PDF para distribui√ß√£o normal
x = np.linspace(-3, 3, 100)
pdf = norm.pdf(x)

plt.subplot(2, 1, 1)
plt.plot(x, pdf)
plt.title('PDF - Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
```

## **CDF (Cumulative Distribution Function)**
* **Defini√ß√£o:** A CDF (Fun√ß√£o de Distribui√ß√£o Acumulada) √© a fun√ß√£o que descreve a probabilidade acumulada at√© um valor espec√≠fico. Ela √© aplic√°vel tanto para vari√°veis discretas quanto cont√≠nuas e fornece a probabilidade de a vari√°vel aleat√≥ria ser menor ou igual a um valor espec√≠fico.
* **Exemplo:** Para o lan√ßamento de um dado, a CDF para o valor 4 √© a soma das probabilidades de obter 1, 2, 3 ou 4.
* **F√≥rmula:** $ùêπ(ùë•)=ùëÉ(ùëã‚â§ùë•)$, onde $ùêπ(ùë•)$ √© a fun√ß√£o de distribui√ß√£o acumulada.

```python
# CDF para distribui√ß√£o normal
cdf = norm.cdf(x)

plt.subplot(2, 1, 1)
plt.plot(x, cdf)
plt.title('CDF - Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.ylabel('Probabilidade Acumulada')
```

## **SF (Survival Function)**
* **Defini√ß√£o:** A SF (Fun√ß√£o de Sobreviv√™ncia) √© a complementa√ß√£o da CDF. Ela representa a probabilidade de a vari√°vel aleat√≥ria ser maior que um determinado valor. Tamb√©m conhecida como fun√ß√£o de confiabilidade, √© amplamente utilizada em an√°lise de sobreviv√™ncia e confiabilidade.
* **Exemplo:** Se voc√™ est√° estudando a vida √∫til de um dispositivo, a SF pode dizer a probabilidade de o dispositivo durar mais do que um certo n√∫mero de horas.
* **F√≥rmula:** $ùëÜ(ùë•)=1‚àíùêπ(ùë•)$, onde $ùëÜ(ùë•)$ √© a fun√ß√£o de sobreviv√™ncia e $ùêπ(ùë•)$ √© a CDF.

```python
# SF para distribui√ß√£o normal (1 - CDF)
sf = norm.sf(x)

plt.subplot(2, 1, 1)
plt.plot(x, sf)
plt.title('SF - Fun√ß√£o de Sobreviv√™ncia')
plt.xlabel('Valor')
plt.ylabel('Probabilidade de Sobreviv√™ncia')

plt.tight_layout()
plt.show()
```

# **PPF (Percent Point Function)**
* **Defini√ß√£o:** A PPF (Fun√ß√£o do Ponto Percentual) √© a inversa da CDF. Ela retorna o valor de entrada para uma dada probabilidade acumulada. Em outras palavras, para um valor de probabilidade, ela retorna o ponto de percentil correspondente.
* **Exemplo:** Para uma distribui√ß√£o normal, a PPF pode nos dar o valor de z correspondente ao percentil 95%.
* **F√≥rmula:** $ùêπ^{‚àí1}(ùëù)$, onde $ùêπ^{‚àí1}$ √© a inversa da CDF e $ùëù$ √© a probabilidade acumulada.

```python
# PPF para distribui√ß√£o normal
ppf_values = np.linspace(0, 1, 100)
ppf = norm.ppf(ppf_values)

plt.subplot(2, 1, 1)
plt.plot(ppf_values, ppf, 'o')
plt.title('PPF - Pontos Percentis da Distribui√ß√£o Normal')
plt.xlabel('Probabilidade')
plt.ylabel('Valor')
```

# **RVS (Random Variate Sample)**
* **Defini√ß√£o:** A RVS (Amostra Aleat√≥ria de Vari√°veis) gera amostras aleat√≥rias a partir de uma distribui√ß√£o de probabilidade especificada.
* **Exemplo:** Gerar uma amostra de 1000 valores de uma distribui√ß√£o normal com m√©dia 0 e desvio padr√£o 1.
* **F√≥rmula:** `ùëüùë£ùë†(size)`, onde `size` √© o n√∫mero de amostras desejadas.

```python
# RVS para distribui√ß√£o normal
rvs = norm.rvs(size=1000)

plt.subplot(2, 1, 1)
plt.hist(rvs, bins=20, density=True)
plt.title('RVS - Amostra Aleat√≥ria da Distribui√ß√£o Normal')
plt.xlabel('Valor')
plt.ylabel('Densidade')

plt.tight_layout()
plt.show()
```

‚ùáÔ∏è Exemplo com bananas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# Par√¢metros da distribui√ß√£o normal (m√©dia e desvio padr√£o)
mu, sigma = 5, 1

# Valores cont√≠nuos para o consumo de bananas
x = np.linspace(0, 10, 100)

# PMF para distribui√ß√£o binomial simulada
bananas = np.arange(11)
probabilidades = [0.01, 0.02, 0.05, 0.1, 0.2, 0.25, 0.15, 0.1, 0.08, 0.03, 0.01]

plt.subplot(3, 2, 1)
plt.bar(bananas, probabilidades)
plt.title('PMF - N√∫mero de Bananas por Semana')
plt.xlabel('N√∫mero de Bananas')
plt.ylabel('Probabilidade')

# PDF para distribui√ß√£o normal
pdf = norm.pdf(x, mu, sigma)

plt.subplot(3, 2, 2)
plt.plot(x, pdf)
plt.title('PDF - Distribui√ß√£o Normal do Consumo de Bananas')
plt.xlabel('N√∫mero de Bananas')
plt.ylabel('Densidade de Probabilidade')

# CDF para distribui√ß√£o normal
cdf = norm.cdf(x, mu, sigma)

plt.subplot(3, 2, 3)
plt.plot(x, cdf)
plt.title('CDF - Distribui√ß√£o Acumulada do Consumo de Bananas')
plt.xlabel('N√∫mero de Bananas')
plt.ylabel('Probabilidade Acumulada')

# SF para distribui√ß√£o normal (1 - CDF)
sf = norm.sf(x, mu, sigma)

plt.subplot(3, 2, 4)
plt.plot(x, sf)
plt.title('SF - Fun√ß√£o de Sobreviv√™ncia do Consumo de Bananas')
plt.xlabel('N√∫mero de Bananas')
plt.ylabel('Probabilidade de Sobreviv√™ncia')

# PPF para distribui√ß√£o normal
ppf_values = np.linspace(0, 1, 100)
ppf = norm.ppf(ppf_values, mu, sigma)
plt.subplot(3, 2, 5)
plt.plot(ppf_values, ppf, 'o')
plt.title('PPF - Pontos Percentis do Consumo de Bananas')
plt.xlabel('Probabilidade')
plt.ylabel('N√∫mero de Bananas')

# RVS para distribui√ß√£o normal
rvs = norm.rvs(mu, sigma, size=1000)
plt.subplot(3, 2, 6)
plt.hist(rvs, bins=20, density=True)
plt.title('RVS - Amostra Aleat√≥ria da Distribui√ß√£o de Bananas')
plt.xlabel('N√∫mero de Bananas')
plt.ylabel('Densidade')

plt.tight_layout()
plt.show()
```
