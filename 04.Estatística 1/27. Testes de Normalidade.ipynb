# Testes de Normalidade

**O que são testes de normalidade?**

Imagine que você tem um conjunto de dados, como as alturas de um grupo de pessoas. <br>
Antes de aplicar certos testes estatísticos ou construir modelos, é importante verificar se esses dados seguem uma distribuição normal (em forma de sino). <br>
Essa verificação é feita através de testes de normalidade.

**Por que a normalidade é importante?**

Muitos testes estatísticos e modelos assumem que os dados seguem uma distribuição normal. Se essa premissa não for verdadeira, os resultados podem ser enganosos.



**Quais são os testes de normalidade mais comuns em Python?**

* **Teste de Shapiro-Wilk:** Um dos mais utilizados, especialmente para amostras menores.
* **Teste de Kolmogorov-Smirnov:** Comum para amostras maiores, mas sensível a desvios na cauda da distribuição.
* **Teste de Anderson-Darling:** Outro teste poderoso, especialmente sensível a desvios da distribuição normal nos extremos.

```python
# A biblioteca scipy.stats oferece funções para realizar esses testes:
from scipy import stats
from scipy.stats import norm, skewnorm

# Suponha que seus dados estão em uma lista chamada 'dados'
dados = norm.rvs(size = 1000) # RVS = Random Variates of given type./ #Norm = Dados não inviesados próximos da distribuição normal

# Teste de Shapiro-Wilk
stat, p = stats.shapiro(dados)
print('Shapiro-Wilk W = %.4f, p = %.4f' % (stat, p))

# Teste de Kolmogorov-Smirnov
stat, p = stats.kstest(dados, 'norm')
print('K-S statistic = %.4f, p = %.4f' % (stat, p))

# Teste de Anderson-Darling
result = stats.anderson(dados)
print('Anderson-Darling statistic: %.3f' % result.statistic)
```

**Interpretando os resultados:**

* **Hipótese nula:** Os dados seguem uma distribuição normal.
* **P-valor:** Se o p-valor for menor que o nível de significância (comumente 0.05), rejeitamos a hipótese nula e concluímos que os dados não são normais.

**Visualizando a distribuição:**

Além dos testes estatísticos, é útil visualizar a distribuição dos dados. Um histograma e um gráfico Q-Q plot podem ajudar a identificar desvios da normalidade.

```python
import matplotlib.pyplot as plt

# Histograma
plt.hist(dados)
plt.show()

# Q-Q plot
stats.probplot(dados, plot=plt)
plt.show()
```

**O que fazer se os dados não forem normais?**

* **Transformação dos dados:** Algumas transformações (como logarítmica) podem ajudar a normalizar os dados.
* **Uso de testes não paramétricos:** Esses testes não assumem normalidade.
* **Considerar outras distribuições:** Talvez seus dados sigam outra distribuição, como a distribuição t de Student ou a distribuição exponencial.

```python
#SkewNorm = Dados inviesados sem distribuição normal
dados2 = skewnorm.rvs(4, size=1000) 
```

```python
# Histograma
plt.hist(dados2)
plt.show()

stats.probplot(dados2, plot=plt)
plt.show()
```
