```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
dados = norm.rvs(size = 1000) # RVS = Random Variates of given type./ #Norm = Dados não inviesados próximos da distribuição normal
```

```python
plt.hist(dados, bins=20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados)
```

```python
dados2 = skewnorm.rvs(4, size=1000) #SkewNorm = Dados inviesados sem distribuição normal
```

```python
plt.hist(dados2, bins=20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
