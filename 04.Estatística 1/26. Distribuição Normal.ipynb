# Distribuição Normal

**O que é a Distribuição Normal?**

Imagine uma curva em formato de sino. <br>
Essa é a representação visual da distribuição normal, uma das distribuições de probabilidade mais importantes em estatística. <br>
Ela descreve como os dados se distribuem em torno de um valor central (a média), com a maioria dos dados concentrados perto da média e poucos dados nos extremos.

**Características chave:**

* **Simetria:** A curva é simétrica em relação à média.
* **Forma de sino:** A forma da curva é característica.
* **Média, mediana e moda são iguais:** O ponto mais alto da curva corresponde à média, mediana e moda.
* **Desvio padrão:** Define a largura da curva. Quanto maior o desvio padrão, mais espalhada a distribuição.

**Por que a Distribuição Normal é Importante?**
* **Frequente na natureza:** Muitos fenômenos naturais seguem a distribuição normal (altura das pessoas, notas de provas, erros de medição).
* **Base para muitos testes estatísticos:** Muitos testes estatísticos assumem que os dados seguem uma distribuição normal.
* **Utilizada em diversas áreas:** Engenharia, ciências sociais, finanças, entre outras.


**Bibliotecas Essenciais:**

* NumPy: Para criar arrays e realizar operações numéricas.
* SciPy: Contém módulos para estatística, incluindo funções para a distribuição normal.
* Matplotlib: Para visualização de dados.


**Funções Principais: `scipy.stats.norm`**
* `pdf(x, loc=0, scale=1)`: Calcula a densidade de probabilidade em um dado ponto x, para uma distribuição normal com média loc e desvio padrão scale.
* `cdf(x, loc=0, scale=1)`: Calcula a função de distribuição cumulativa em um dado ponto x.
* `ppf(q, loc=0, scale=1)`: Calcula o quantil para uma dada probabilidade q.
* `rvs(size=1, loc=0, scale=1)`: Gera números aleatórios de uma distribuição normal.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Criar uma distribuição normal com média 50 e desvio padrão 10
mu, sigma = 50, 10
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)

# Plotar a curva
plt.plot(x, y)
plt.xlabel('Valor')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Normal')
plt.show()
```

#### ❇️ Exemplo: Imagine um conjunto de objetos em uma cesta: **média = 8**, **desvio padrão = 2** <br>

```python
from scipy.stats import norm
```

Qual a probabilidade de tirar um objeto com peso menor que 6 Kg? ↓

```python
menor = norm.cdf(6, 8, 2)
print("A probabilidade de tirar um objeto com peso MENOR que 6 Kg é de", round(menor, 2),"%") #Cumulative Distribution Function of the given RV.
```

Qual a probabilidade de tirar um objeto com peso maior que 6 Kg? ↓

```python
maior = norm.sf(6, 8, 2) #Survival function (1 - cdf) ↓ at x of the given RV.
#1 - norm.cdf(6, 8, 2)


print("A probabilidade de tirar um objeto com peso MAIOR que 6 Kg é de", round(maior, 2),"%") #Cumulative Distribution Function of the given RV.
```

Qual a probabilidade de tirar um objeto com peso menor que 6 Kg OU maior que 10 Kg? ↓

```python

ou = norm.cdf(6, 8, 2) + norm.sf(10, 8, 2)

print("A probabilidade de tirar um objeto com peso MENOR que 6 Kg OU MAIOR que 10 Kg é de", round(ou, 2),"%") #Cumulative Distribution Function of the given RV.
```

Qual a probabilidade de tirar um objeto com peso menor que 10 Kg E maior que 8 Kg? ↓

```python
e = norm.cdf(10, 8, 2) - norm.cdf(8, 8, 2)

print("A probabilidade de tirar um objeto com peso MENOR que 6 Kg OU MAIOR que 10 Kg é de", round(e, 2),"%") #Cumulative Distribution Function of the given RV.
```
